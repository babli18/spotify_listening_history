```python
import spotipy
from dotenv import load_dotenv
import os
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd
```

```python
client_id = os.getenv('CLIENT_ID')
client_secret = os.getenv('CLIENT_SECRET')
uri = os.getenv("URI")

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=client_id,
                                               client_secret=client_secret,
                                               redirect_uri = uri,
                                               scope="user-top-read"))
```

```python
# Get top tracks (long_term = several years of data)
top_tracks_long = sp.current_user_top_tracks(time_range='long_term', limit=50)

# Lists to store data
songs = []
song_ids = []
artists = []

# Extract track info
print("\n--- Top Tracks (Long Term) ---")
for i, track in enumerate(top_tracks_long['items']):
    song_name = track['name']
    song_id = track['id']
    artist_name = track['artists'][0]['name']
    
    print(f"{i+1}. {song_name} by {artist_name}")
    
    songs.append(song_name)
    song_ids.append(song_id)
    artists.append(artist_name)

# Create DataFrame
top_long_term = pd.DataFrame({
    'song': songs,
    'song_id': song_ids,
    'artist': artists
})

top_long_term.to_csv("top_tracks/top_long_tracks.csv", index=False)
```
